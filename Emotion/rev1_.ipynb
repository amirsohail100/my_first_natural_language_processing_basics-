```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes  import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("train.txt",sep=";",header=None,names=["text","emotion"])
df.head()
```

```python
lab_enco = LabelEncoder()
df["emotion"] = lab_enco.fit_transform(df["emotion"])
df.head()
```

```python
df["emotion"].unique()
```

```python
df["text"] = df["text"].apply(lambda X:X.lower())
df.head()
```

```python
import string

def remove_punc(txt):
    return txt.translate(str.maketrans("","",string.punctuation))

df["text"] = df["text"].apply(remove_punc)
df.head()
```

```python
def remove_number(txt):
    new = ""
    for i in txt:
        if not i.isdigit():
            new = new+i
    return new
df["text"] = df["text"].apply(remove_number)
df.head()
```

```python
def remove_enojis(txt):
    new = ""
    for i in txt:
        if i.isascii():
            new+=i
    return new

df["text"] = df["text"].apply(remove_enojis)
df.head()
```

```python
from nltk.corpus import stopwords
s_w = set(stopwords.words("english"))
s_w
```

```python
def remove_stopwords(txt):
    words_list = txt.split()
    cleaned = []
    for i in words_list:
        if not i in s_w:
            cleaned.append(i)
    return ' '.join(cleaned)

df["text"] = df["text"].apply(remove_stopwords)
df.loc[1]["text"]
```

```python
X = df["text"]
y = df["emotion"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
Count_Vectorizer = CountVectorizer()
Tfidf_Vectorizer = TfidfVectorizer()

X_train_Count = Count_Vectorizer.fit_transform(X_train)
X_test_Count = Count_Vectorizer.fit_transform(X_test)
X_train_Tfidf = Tfidf_Vectorizer.transform(X_train)
X_test_Tfidf = Tfidf_Vectorizer.transform(X_test)
```

```python
models = {
    "nb2":MultinomialNB(),
    "LR":LogisticRegression()
}

# X_trains = {
#     "X_train_Count":X_train_Count,
#     "X_train_Tfidf":X_train_Tfidf
# }
# for name,model in models.items():
#     for X_,X_tr in X_trains.items():
#         if X_ == "X_train_Count":
#             model.fit(X_tr,y_train)
#             y_pred = model.predict(X_test_Count)
#             print(f"{name} : {X_}",accuracy_score(y_test,y_pred))
#         else:
#             model.fit(X_tr,y_train)
#             y_pred = model.predict(X_test_Tfidf)
#             print(f"{name} : {X_}",accuracy_score(y_test,y_pred))


for name,model in models.items():
    model.fit(X_test_Count,y_train)
    y_pred = model.predict(X_test_Count)
    print(f"{name} : X_test_Count",accuracy_score(y_test,y_pred))
for name,model in models.items():
    model.fit(X_test_Tfidf,y_train)
    y_pred = model.predict(X_test_Tfidf)
    print(f"{name} : X_test_Tfidf",accuracy_score(y_test,y_pred))
```
