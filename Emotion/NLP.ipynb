```python
from sklearn.feature_extraction.text import CountVectorizer
doc = [
    "I love pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(doc)

print(f"Vectorizer : {vectorizer.get_feature_names_out()}")
print(f"\nBow Matrtx : \n {X.toarray()}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
doc = [
    "I love pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]

vectorizer = CountVectorizer(ngram_range=(2,2))

X = vectorizer.fit_transform(doc)

print(f"Vectorizer : {vectorizer.get_feature_names_out()}")
print(f"\nBow Matrtx : \n {X.toarray()}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
doc = [
    "I love pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(doc)

print(f"Vectorizer : {vectorizer.get_feature_names_out()}")
print(f"\nBow Matrtx : \n {X.toarray()}")
```
